```python
import numpy as np
import math
import scipy
import matplotlib.pyplot as plt

#EPS Foam Compression regression from reseach gate data

densities = np.array([43, 64, 80, 120])
yields = np.array([0.33, 0.53, 0.70, 1.26])
initial = np.array([10,-2])
def sumedSquares(x):
    a0 = x[0]
    a1 = x[1]
    E = np.sum((np.log(yields) - (a1*np.log(densities) + a0))**2)
    return E

finals = scipy.optimize.minimize(sumedSquares, initial)
print(finals.x)


print(np.log(densities))
print(np.log(yields))
density = np.linspace(0,5,num=250)
yieldStress = finals.x[1]*(density) + finals.x[0]

plt.scatter(density, yieldStress)
plt.scatter(np.log(densities), np.log(yields))
```
